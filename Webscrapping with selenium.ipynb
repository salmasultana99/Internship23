Q2: Write a python program to scrape data for ‚ÄúData Scientist‚Äù Job position in ‚ÄúBangalore‚Äù location. You

This task will be done in following steps:
1. First get the webpage https://www.naukri.com/
2. Enter ‚ÄúData Scientist‚Äù in ‚ÄúSkill, Designations, Companies‚Äù field and enter ‚ÄúBangalore‚Äù in ‚Äúenter the
location‚Äù field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

```python

#getting Url
url = 'https://www.naukri.com/'
driver.get(url)

search_job = driver.find_element_by_xpath("//input[@placeholder='Enter skills / designations / companies']") #search bar
search_job.send_keys("Data Scientist") #Adding "Data Scientist" in search bar
search_loc = driver.find_element_by_xpath("//input[@placeholder='Enter location']") #search bar
search_loc.send_keys("Banglore") #Adding "Banglore" in search bar

search_btn = driver.find_element_by_xpath("//div[@class='qsbSubmit']") #submit button
search_btn.click()

time.sleep(2) #Sleep, just to make sure page will load perfectely 

titles_tags=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
job_titles=[]
for i in range(0,10):
    job_titles.append(titles_tags[i].text)
    
companies_tags=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
company_names=[]
for i in range(0,10):
    company_names.append(companies_tags[i].text)

locations_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span")
locations_list=[]
for i in range(0,10):
    locations_list.append(locations_tags[i].text)
    
job_url=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
urls=[]
for i in range(0,10):
    urls.append(job_url[i].get_attribute('href'))
    
description_list=[]
for i in urls:
    driver.get(i)
    try:                  #exception handling for nosuchelementexception
        description_tags=driver.find_element_by_xpath("//div[@class='dang-inner-html']")#locating the description
        description_list.append(description_tags.text.replace("\n",' '))#appending the description in description list
    except NoSuchElementException as e:
        try:                                                      #click the rating link found
            description_tags=driver.find_element_by_xpath("//div[@class='nConfig_textblock ']")
            description_list.append(description_tags.text.replace("\n",' '))
        except NoSuchElementException as e:
            description_list.append("Not found")#appending not found if no description is there
            
jobs=pd.DataFrame({})
jobs['Title'] = job_titles
jobs['Company'] = company_names
jobs['Description'] = description_list
jobs['Location'] = locations_list
jobs
```

Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
1. Brand
2. Product Description
3. Price
To scrape the data you have to go through following steps:
1. Go to Flipkart webpage by url : https://www.flipkart.com/
2. Enter ‚Äúsunglasses‚Äù in the search field where ‚Äúsearch for products, brands andmore‚Äù is written and
click the search icon
3. After that you will reach to the page having a lot of sunglasses. From this pageyou can scrap the
required data as usual.
4. After scraping data from the first page, go to the ‚ÄúNext‚Äù Button at the bottom ofthe page , then
click on it.
5. Now scrape data from this page as usual
6. Repeat this until you get data for 100 sunglasses.

```python
url = 'https://www.flipkart.com/'
driver.get(url)

btn = driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']") #closing login popup
btn.click()

search_job = driver.find_element_by_class_name("_3704LK") #search bar
search_job.send_keys("sunglasses")

search_btn = driver.find_element_by_xpath("//button[@class='L0Z3Pu']") #search button
search_btn.click()

time.sleep(4)

Brand=[]
Description=[]
Price=[]
Discount=[]
for i in range(0,3):
    brand_name=driver.find_elements_by_xpath("//div[@class='_4ddWXP']/a[2]")
    for i in brand_name:
        Brand.append(i.text)
    desc=driver.find_elements_by_xpath("//div[@class='_4ddWXP']/div[1]")
    for i in desc:
        Description.append(i.text)
    glass_price=driver.find_elements_by_xpath("//div[@class='_30jeq3']")
    for i in glass_price:
        Price.append(i.text)
    prod_disc=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']/span")
    for i in prod_disc:
        Discount.append(i.text)
    
    nxt=driver.find_elements_by_xpath("//a[@class='_1LKTO3']")
    url=[]
    for i in nxt:
        url=i.get_attribute('href')
    time.sleep(2)
    driver.get(url)
    
Glass=pd.DataFrame({})
Glass['Brand'] = Brand
Glass['Description'] = Description
Glass['Price'] = Price
Glass['Discount'] = Discount
Glass
```
